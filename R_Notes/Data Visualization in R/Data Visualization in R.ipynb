### Histograms

```r
set.seed(145)
a=rnorm(1000,7,5)
b=rnorm(1000,7,6)
ab=data.frame(a,b)

par(mfrow = c(3,3))
# simple histogram
hist(a) 
# if you just want a numeric output of the hist statistics
hist(a, plot=F) 

# if you want a histogram without axes and annotations (but with labels) 
hist(a, axes=F, ann=F, labels=T, ylim=c(0,220))

# ways to adjust the bin size
hist(a, breaks=c(-7,3,6,7,9,15,20,25))

hist(a, breaks=10)

# we can put the labels in the plot (watch ylim)
hist(a, breaks=10, col=c("forestgreen","skyblue"), 
     main="Random Histogram", col.main="grey33", 
     cex.main=1.3, ylab="Freq", xlab="",
     labels=T, ylim=c(0,450), las=3)
mtext("base",cex=0.7, side=3)

# we can put the output to density
hist(a, freq=F, xlab="", main="Density plot")

# let's put 2 histograms in 1 plot
h1 = hist(a, breaks=c(-15,-10,-5,0,5,10,15,20,25,30))
h2 = hist(b, breaks=c(-15,-10,-5,0,5,10,15,20,25,30))

# we can use alpha blending to overlap the histograms
plot(h1, col=rgb(1,0,0,0.25), ylim=c(0,410), xlab="", 
     main="Combined Histogram")
plot(h2, col=rgb(0,1,0,0.25), ylim=c(0,410),add=T)
mtext("base",cex=0.7, side=3)
legend(20,300, legend=c("a","b"), bty="n", text.col=c("red", "green"))
```

### 3D pie chart

+  plotrix offers the possibility to draw a 3D piechart with explode effect 
+  height defines the height of the chart - default is good
+  theta defines the angle in which the chart is displayed
+  border = line coloration
+  shade = shade effect between 0 and 1
+  explode effect should be around 0.05 - 0.1  

```r
values=c(35,35,30)
labels=c("India", "China", "Thailand")
pie3D(x=values, labels=labels,
      height=0.2, theta=pi/4, border="brown", 
      shade=0.5, explode=0.07, main="3D Pie Chart")
mtext("plotrix",cex=0.7, side=3)
```

### Gantt Chart - organizational charts, time management

+ plotrix offers a simple gantt chart.
+ gantt.chart
+ at first we need to set our time format 

```r
Ymd.format <- "%Y/%m/%d" 
# we set our tasks, start and end points using POSIXit format
tasks <- list(labels= c("Research","Plan","Buy Material","Produce","Market"),
              starts= as.POSIXct(strptime( c("2011/01/11","2011/03/11","2011/06/13","2011/09/16","2012/01/12"), format=Ymd.format)),
              ends= as.POSIXct(strptime(c("2011/04/17","2011/07/18","2011/08/22","2011/12/06","2012/11/12"), format=Ymd.format)),
              priorities=c(2,2,2,1,1)) # even priorities are possible

# now we need to specify the x axis (time axis)

# this fixes the positions of the months vector
posofmonths <- as.POSIXct(strptime(c("2011/01/11", "2011/09/11", "2012/05/11", "2012/11/12"), format=Ymd.format))

# the time for our x scale
months <- c("2011 Jan", "2011 Sep", "2012 May", "2012 Nov") 
# actual chart code
gantt.chart(tasks, main="Gantt Chart - plotrix", 
            priority.legend=TRUE,
            vgridlab=months, 
            vgridpos=posofmonths,
            hgrid=T) # hgrid for orientation dots
```

### Gap plots

+ can be used if wide ranges on axes are left without values (alternative to logarithm)
