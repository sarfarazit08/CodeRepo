# Recursive Functions

```r
# Fibonacci series

fib <- function(n){
    if (n>2){
        c(fib(n-1), sum(tail(fib(n-1),2)))
    }
    else if(n>=0){
        rep(1,n)
    }
}
    
fib(1)

fib(2)

fib(10)
```

* Recusrsion could be slow and so use it when performance is not an issue.

```r
system.time(fib(20))
```

** a much faster version **

```r
# Fibonacci series (but only gives single output)

fib <- function(n){
    if (n==0){
        return (0)
    }
    if (n==1){
        return (1)
    }
    return (fib(n-1)+fib(n-2))
}

system.time(fib(20))
```

```r
# Factorial

factorial <- function(n){
    if (n==1){
        return (1)
    }else {
        return (n * factorial(n-1))        
    }
    
}
factorial(10)
system.time(factorial(20))
```
