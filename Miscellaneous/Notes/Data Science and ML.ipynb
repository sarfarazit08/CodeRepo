<h3>Mean vs. Median</h3>

Let's create some fake income data, centered around 30,000 
with a normal distribution and standard deviation of 15,000 with 10,000 data points.

Then, compute the mean(average) - it should be close to 30,000 :

```python
import numpy as np

incomes = np.random.normal(30000, 15000, 10000)
np.mean(incomes)
```

we can segment the income data into 100 buckets, and plot it as a histogram.

```python
import matplotlib.pyplot as plt

plt.hist(incomes,100)
plt.show()
```

Now compute the median - since we've a nice, even distribution it too should be close to 30000.

```python
np.median(incomes)
```

Now we'll add Bill Gates income to the mix to show darn income inequality!

```python
incomes = np.append(incomes,[1000000000])
```

Median won't change much, but the mean does:

```python
np.median(incomes)
```

```python
np.mean(incomes)
```

**The above experiment shows that 'MEAN' is going to change incase of outliers in the data.**

<h1>Moments : Mean, Variance, Skew, Kurtosis</h1>

Create a roughly normal-distributed random set of data.

```python
import numpy as np
import matplotlib.pyplot as plt

vals = np.random.normal(0, 0.5, 10000)
plt.hist(vals, 50)
plt.show()
```

**First Moment**

The first moment is the _Mean_. This data should average out to be 0 (zero).

```python
np.mean(vals)
```

**Second Moment**

The second moment is the _Variance_.

```python
np.var(vals)
```

 **Third Moment**

The third moment is the _Skew_. Since data is nicely centered around 0(zero), it should be almost zero(0).

```python
import scipy.stats as sp

sp.skew(vals)
```

**Fourth Moment**

The fourth moment is the _Kurtosis_. It describes the shape of the tail. For a normal disribution it is zero(0).

```python
sp.kurtosis(vals)
```

### Covariance and Correlation 

Covarance measures how two variabls vary in tandem from their means.

For example, let's say we work for an e-commerce company, and they are interested in finding a correlation between
page speed (how fast each web page returned to customer) and how much a customer spends.

**numpy** offers covariance methods, but we'l do it the "hard way" to see, what happens under the hood. 

Basically, we treat each varable as a vector of deviations from the mean, and compute the "dot product" of both vectors. 

Geometrcaty this can be thought of as the angle between the two vectors in a high-dimensional space, 
but you can just think of it as a measure ot similarity between the two variables.

First, let's just make page speed and purchase amount totally random and independent of each other;
a very small covariance will resut as there is no real correlation.

```python
import numpy as np
import matplotlib.pyplot as plt

def fn_mean(x):
    xmean = np.mean(x)
    return [xi - xmean for xi in x]
def fn_covariance(x, y):
    n=len(x)
    return np.dot(fn_mean(x), fn_mean(y)) / (n-1)

pageSpeed = np.random.normal(3.0,1.0,1000)
purchaseAmount = np.random.normal(50.0, 10.0,1000)

plt.scatter(pageSpeed, purchaseAmount)
plt.show()

fn_covariance(pageSpeed, purchaseAmount)
    
```

Now we'll make our fabricated purchase amounts as an actual function of page speed, making a very real correlation.

The negative calue indicates an inverse relationship; pages that render in less time result in more money spent.

```python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeed

plt.scatter(pageSpeed, purchaseAmount)
plt.show()

fn_covariance(pageSpeed, purchaseAmount)
```

But, what does this value mean? Covariance is sensititve to the units used in the variables, which makes it difficult to interprete.
Correlation normalizes everything by their standard deviations, giving you an easier to understand value that ranges 
from -1 (for a perfect inverse correlation)  to 1 (for a perfect positiv correlation).

```python
import scipy as sp

def correlation(x, y): 
    stddevx = x.std()
    stddevy = y.std()
   
    try :
        return fn_covariance(x,y) / stddevx / stddevy 
    except:
        print("Something went wrong.")

correlation (pageSpeed, purchaseAmount) 
```

numpy can do all this for you with numpy.corrcoef. It returns a matrix of the correlation coefficients between every combinabon of the arrays passed in: 

```python
import numpy as np
np.corrcoef(pageSpeed, purchaseAmount) 
```

(It doesn't match exactly just due to the math precson avaialable on a computer.)

We can force a perfect correlation by fabricating a totally linear relationship (again, it's not exactly -1 just due to precision errors, but it's close enough to tell us there's a really good correlation here):

```python
purchaseAmount = 100 - pageSpeed * 3

plt.scatter(pageSpeed, purchaseAmount)
plt.show()

correlation(pageSpeed, purchaseAmount)
```

**_Remember, correlation does not imply causality!_**
